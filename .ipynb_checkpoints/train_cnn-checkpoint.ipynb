```python
import os
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt
```

```python
# Path to the labels CSV
labels_path = 'labels.csv'  # Replace with your actual path to labels.csv

# Load the labels CSV
labels_df = pd.read_csv(labels_path)

# Check the contents of the labels file to ensure it's loaded correctly
labels_df.head()
```

```python
# Path to your image data folder
data_dir = 'myData'  # Replace with the path to your images folder

# Prepare the images and their corresponding labels
images = []
labels = []

for index, row in labels_df.iterrows():
    folder_path = os.path.join(data_dir, str(row['ClassId']))  # Folder is named by ClassId
    for file_name in os.listdir(folder_path):
        # Skip hidden files or directories (e.g., .ipynb_checkpoints)
        if file_name.startswith('.') or not file_name.endswith(('.png', '.jpg', '.jpeg')):
            continue
        image_path = os.path.join(folder_path, file_name)
        try:
            image = load_img(image_path, target_size=(64, 64))  # Resize image to 64x64
            image = img_to_array(image) / 255.0  # Normalize image pixels
            images.append(image)
            labels.append(row['ClassId'])
        except Exception as e:
            print(f"Error loading image {image_path}: {e}")

# Convert to numpy arrays
X = np.array(images)
y = np.array(labels)

# Encode the labels (ClassId) into integers
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Check the shape of the data
print(f'Images shape: {X.shape}')
print(f'Labels shape: {y.shape}')
```

```python
# Define the CNN model architecture
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),  # First convolutional layer
    MaxPooling2D(pool_size=(2, 2)),  # MaxPooling layer
    Conv2D(64, (3, 3), activation='relu'),  # Second convolutional layer
    MaxPooling2D(pool_size=(2, 2)),  # MaxPooling layer
    Flatten(),  # Flatten the output from the convolutional layers
    Dense(128, activation='relu'),  # Fully connected layer
    Dense(len(np.unique(y)), activation='softmax')  # Output layer for number of classes
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint

# Save the best model during training using ModelCheckpoint
checkpoint = ModelCheckpoint('best_model.keras', save_best_only=True, monitor='val_loss', mode='min')

# Train the model (you can adjust epochs, batch size, etc.)
model.fit(X, y, epochs=10, validation_split=0.2, batch_size=32, callbacks=[checkpoint])
```

```python
# Plot accuracy and loss during training

# Visualize the training history (accuracy and loss)
def plot_training_history(history):
    # Accuracy plot
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    # Loss plot
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.show()

# Call the function to plot the history
plot_training_history(model.history)

# Print the final accuracy percentage
final_accuracy = model.history.history['accuracy'][-1] * 100
print(f"Final Training Accuracy: {final_accuracy:.2f}%")
```

```python
# Test the model with a new image
test_image_path = 'path_to_your_test_image.jpg'  # Provide the image path here
test_image = load_img(test_image_path, target_size=(64, 64))
test_image_array = img_to_array(test_image) / 255.0  # Normalize image pixels
test_image_array = np.expand_dims(test_image_array, axis=0)  # Add batch dimension

# Predict the class of the image
predicted_class = model.predict(test_image_array)

# Get the predicted class index and corresponding label
predicted_class_index = np.argmax(predicted_class)
predicted_label = label_encoder.inverse_transform([predicted_class_index])

# Print class index and corresponding label to debug
print(f"Predicted class index: {predicted_class_index}")
print(f"Predicted class label: {predicted_label[0]}")

# Convert the image for displaying (from PIL to NumPy array)
test_image_pil = Image.open(test_image_path)

# Add predicted label to the image
draw = ImageDraw.Draw(test_image_pil)
label = f"Predicted: {predicted_label[0]}"
font = ImageFont.load_default()  # Use default font
draw.text((10, 10), label, fill=(255, 0, 0), font=font)

# Convert back to NumPy array for plotting
test_image_np = np.array(test_image_pil)

# Show the image with the predicted label
plt.imshow(test_image_np)
plt.axis('off')  # Turn off axis
plt.show()

# Print the predicted label in the console
print(f"Predicted class: {predicted_label[0]}")
```
